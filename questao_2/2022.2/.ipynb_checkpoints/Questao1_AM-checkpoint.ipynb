### Resolução do Algoritmo RBSOM-CWMdd com 50 iterações aplicadas ao dataset HTRU2.txt [Candidatos a pulsar]

```python
#Passo 1: Matriz de Dissimilaridade
def getMatrizDissimilaridadeNormalizada(dataset):
    dados = dataset.iloc[:,0:8]

    dissimilaridade = pd.DataFrame(distance_matrix(dados.values,dados.values),index = dados.index,columns=dados.index)

    norm_diss = np.array(
        dissimilaridade
        / dissimilaridade.iloc[np.argmin(dissimilaridade.sum())].sum()
    )
    
    return norm_diss
```

```python
matrizDissimilaridadeNormalizada = getMatrizDissimilaridadeNormalizada(dataset)

matrizDissimilaridadeNormalizada
```

```python
#Passo 1.1: Tamanho do mapa (SIZEMAP - GRID)
x = 3
y = 3

# Passo 1.2: Definição de C (número de neurons);
C = int(x*y)
```

```python
#Passo 2: Matriz de distâncias quadráticas entre nós do grid d

#Criação do Grid que é um array de coordenadas (regiões)
grid = []
for i in range(x):
    for j in range(y):
        grid.append([i,j])

#print(grid)        

#Matriz de distancias quadradas 
squaredDistance = np.empty((C,C))

for i in range(squaredDistance.shape[0]):
    for j in range(squaredDistance.shape[1]):
        squaredDistance[i,j] = np.square(distance.euclidean(grid[i],grid[j])) #Uso de distância euclidiana

print(squaredDistance)
```

```python
#Passo 3: Definições de Niter (número de iterações); sigma_inicial (raio inicial); sigma_final (raio final); q (cardinal do set-medoids)

N_iter = 50;
q = 5  #A cardinalidade do conjunto de medoids
n = 1.1 #Valor padrão para calcular 

sigma_inicial = np.sqrt(-(x * x + y * y) / (2 * np.log(0.1)))
sigma_final = np.sqrt(-1 / (2 * np.log(0.01)))
print('Sigma_Inicial', sigma_inicial, '\nSigma_final', sigma_final)
```

```python
#Passo 4: criação do mapa SOM; vetor de set-medoids G; matriz de relevância dos pesos V; partição p
som = None;
V = []; #matriz de relevância dos pesos V
G = []; #vetor de set-medoids G
p = []; #partição p
```

```python
#Função auxiliar para calcular o raio de acordo com o t

def calcularSigma(t):
   return sigma_inicial*pow((sigma_final/sigma_inicial),t/N_iter)
```

```python
#Função auxiliar para calcular a função de Kernel da vizinhança

def calcularMatrizH(s, r, sigma):
    s = int(s)
    r = int(r)
    dist = squaredDistance[s, r]
    return np.exp(-dist / (2 * np.square(sigma)))
```

```python
#Passo 5: Initialization
t = 0;

#Passo 6: Calcular a matriz H de acordo com a Equação 3 do artigo.

sigma_atual = calcularSigma(t);

matrizH = np.empty((C,C))

for i in range(matrizH.shape[0]):
    for j in range(matrizH.shape[1]):
        matrizH[i,j] = calcularMatrizH(i,j,sigma_atual)

print(matrizH)        
```

```python
#Passo 7: Definição do cluster iniciail de representativos: seleção randômica de C set-medoids distintos para obter o vetor inicial de set-medoids

#G set-medoids: Conjunto de elementos que compoem o prototipo
G= []  
for i in range(C):
    G.append(dataset.sample(q).index); #Indices das linhas do dataset, Em cada linha tem 5 itens, sendo preenchido aleatoriamente
G=np.array(G)

G #impressao dos set-medoids, cada linha representa um set-medoid
```

```python
# Função auxiliar para cálculo da da matriz de pesos

def calcularMatrizDePesosDeRelevancia():
    matrizPesos = pd.DataFrame(np.random.rand(C,q)) #Gera uma matriz/vetor e 9x5 
    somaMatrizPesos = np.array(matrizPesos.sum(axis=1)) # A soma dos elementos de cada grupo de medoids[cada linha] será igual a 1
    matrizPesos = np.array(matrizPesos)

    for i in range(matrizPesos.shape[0]):
      for j in range(matrizPesos.shape[1]):
        matrizPesos[i,j] = matrizPesos[i,j]/somaMatrizPesos[i]

    return matrizPesos;
```

```python
#Passo 8: Calcular a matriz inicial de pesos de relevância de forma randômica
#Vetor de Pesos associado comos medoids dos protótipos correspondentes
V = []

V = calcularMatrizDePesosDeRelevancia()

print(V)
```

```python
#A função Dissimilaridade(Ek,Gr), parametrizado pelo vetor de pesos V, 
#calcula a correspondência entre um objeto eK e um representante do cluster Gr na dissimilaridade D
def calcularCorrespondenciaObjetoEeClusterNaDissimilaridade(k,r):
    return sum(
        [
            np.power(V[r, i], n) * matrizDissimilaridadeNormalizada[k, G[r, i]] #Elementos da primeira matriz elevados a potências da segunda matriz, elemento a elemento.
            for i in range(V.shape[1])
        ]
    )


#A BMU é indexada por f(ek) e identificado com o protótipo Gf(ek). A medida de erro de uma BMU Gf(ek) em relação ao objeto ek 
#é calculado por meio da função correspondente Δ que permite comparar cada objeto ek para o protótipo Gf(ek) de acordo com:
#(2)Δ(ek,Gf(ek))=Σr=1C hf(ek),rD(ek,Gr) Onde hf(ek),r é a função kernel da vizinhança : mede a vizinhança de influência da BMU no neurônio r.
def calcularFuncaoCorrespondenteObjetoParaOPrototipo(k, s):
    return sum([matrizH[s, r] * calcularCorrespondenciaObjetoEeClusterNaDissimilaridade(k, r) for r in range(C)])

##def get_BMU():
```

```python
#Passo 9: Passo de initial assignment para obter a partição inicial que é atualizada de acordo com Eq.(11) do artigo
for i in range(C):
  p.append([]);

bmu = np.zeros(dataset.shape[0])
bmusProximos = np.zeros(dataset.shape[0])

for k in range(dataset.shape[0]):
    deltas = np.array([calcularFuncaoCorrespondenteObjetoParaOPrototipo(k, s) for s in range(C)])
    k_cluster = np.argmin(deltas)
    bmu[k] = k_cluster
    bmusProximos[k] = np.argsort(deltas)[1]
    p[k_cluster].append(k)

bmu = bmu.astype(int)

print(bmu)
print(bmusProximos)
print(deltas)
```

```python
# Função auxiliar para cálculo do quantization error

def CalcularQuantizationError (N, deltas): #N=dataset.shape[0]
  MediaQuantizationError = deltas/N
  DesvioPadQuantizationError = deltas.std()
  return MediaQuantizationError, DesvioPadQuantizationError  
```

```python
# Função auxiliar para cálculo do topographical error

def CalcularTopographicalError (N, bmus, bmusProximos, squaredDistance): #N=dataset.shape[0]
      
  bmus = bmus.astype(int)
  bmusProximos = bmusProximos.astype(int)

  erroTopografico = np.sum([squaredDistance[bmus[k], bmusProximos[k]] == 1 for k in range(N)])
  
  MediaErroTopografico = erroTopografico/N
  DesvioErroTopografico = erroTopografico.std()

  return MediaErroTopografico, DesvioErroTopografico;
```

```python
#Passo 10: Treinamento
loss = []
listaMediaErroTopografico = []
listaDesvioErroTopografico = []
listaMediaErroQuantization = []
listaDesvioErroQuantization = []

P = [[] for i in range(C)]
g = []

#Passo 11: setar objetos da matrizH
for t in tqdm(range(1,N_iter+1)):
    
    #otimizar com um garbage colector
    gc.collect()

    #Atualização do raio de vizinhança de acordo com t da iteração 
    sigma = calcularSigma(t)
    
    
    for i in range(C):
        for j in range(C):
            matrizH[i,j] = calcularMatrizH(i,j,sigma)
    
    #Passo 12
    ## step 1: Representation: Cálculo dos elemento de Gr no vetor de set-medoids
    for r in range(C):
        #g = []

        for h in range(dataset.shape[0]):
            hfec = np.array([matrizH[x,r] for x in bmu])
            dist = matrizDissimilaridadeNormalizada[h]
            val = np.dot(hfec, dist)

            g.append((h,val))
            
        g = sorted(g, key= lambda x: x[1])[:q]
        
        G[r] = np.array([pair[0] for pair in g])
    
    
    ## step 2: Weighting: Cálculo dos elementos de peso de acordo com a equação 10
    for r in range(V.shape[0]): #varre todas as linhas de V
        for e_idx in range(V.shape[1]): #varre todas as colunas de V
            base_sum = []

            upper = sum(
                [
                    matrizH[bmu[k],r]
                    *matrizDissimilaridadeNormalizada[k,G[r,e_idx]]
                    for k in range(dataset.shape[0])
                ]
            )
            
            for element in G[r]:
                lower = np.sum(
                    [
                        matrizH[bmu[k],r]
                        *matrizDissimilaridadeNormalizada[k,element]
                        for k in range(dataset.shape[0])
                    ]
                )

                base_sum.append(
                    pow(upper/lower,(1/(n-1)))
                )
            
            V[r][e_idx] = 1/sum(base_sum)
            
    ## step 3: Assignment: obtêm a nova partição de acordo com a equação 11
    
    for k in range(dataset.shape[0]):
        deltas = np.array([calcularFuncaoCorrespondenteObjetoParaOPrototipo(k, s) for s in range(C)])
        k_cluster = np.argmin(deltas)
        bmu[k] = k_cluster
        bmusProximos[k] = np.argsort(deltas)[1]
        P[k_cluster].append(k)
    
    bmu = bmu.astype(int)
    bmusProximos = bmusProximos.astype(int)
        
    ## Cálculo de loss
    loss_t = sum([calcularFuncaoCorrespondenteObjetoParaOPrototipo(k, bmu[k]) for k in range(dataset.shape[0])])
    loss.append(loss_t)
    print(t,'----', loss_t)

    ## Cálculo do QE e TE
    mediaQErro, desvioQErro = CalcularQuantizationError (dataset.shape[0], deltas);
    mediaTErro, desvioTErro = CalcularTopographicalError (dataset.shape[0], bmu, bmusProximos, squaredDistance);

    ## Lista de QE
    listaMediaErroQuantization.append(mediaQErro);
    listaDesvioErroQuantization.append(desvioQErro);

    #Lista de Erros Topográficos
    listaMediaErroTopografico.append(mediaTErro);
    listaDesvioErroTopografico.append(desvioTErro);
    
    mean = np.mean(mediaQErro)
    
    #melhores valores
    loss_ref = 10
    if(loss_t < loss_ref):
      loss_ref = loss_t
      melhor_t = t
      melhor_erro = loss_t
      melhor_QE_media = listaMediaErroQuantization[melhor_t-1]
      melhor_QE_desvio = listaDesvioErroQuantization[melhor_t-1]
      melhor_TE_media = listaDesvioErroQuantization[melhor_t-1]
      melhor_TE_desvio = listaDesvioErroQuantization[melhor_t-1]
      melhor_G = G
      melhor_P = P
      melhor_V = V
      melhor_bmu = bmu
```

```python
#Calculo das  Métricas para melhor resultado segundo a função Objetivo
print('MediaQuantizationErro:',np.max(listaMediaErroQuantization))
print('DesvioQuantizationErro:', np.max(listaDesvioErroQuantization))
print('MediaTopograficErro:', np.max(listaMediaErroTopografico))
print('DesvioTopograficErro:', np.max(listaDesvioErroTopografico))
print('\n')
```

```python
# Função auxiliar de criação de combinações
def GeraCombinacoes(n):
    r=2
    if (n-r)<0:
        return 0
    return factorial(n)/(factorial(r)*factorial(n-r))
```

```python
# Cálculo para encontrar IndiceRand
n_i = MatrizContingencia.sum(axis=1)
n_j = MatrizContingencia.sum(axis=0)
M = GeraCombinacoes(MatrizContingencia.sum())

C_n_i = []
C_n_j = []
C_n = []

for i in range(len(MatrizContingencia)):
  C_n_i.append(GeraCombinacoes(n_i[i]))

for j in range(len(n_j)):
  C_n_j.append(GeraCombinacoes(n_j[j])) 

m1 = sum(C_n_i)
m2 = sum(C_n_j)

for i in range(len(n_i)):
  for j in range(len(n_j)):
    C_n.append(GeraCombinacoes(MatrizContingencia[i][j]))

m = sum(C_n)
```

```python
IndiceRandCalculado= (m - (m1*m2)/M)/((m1/2) + (m2/2) - (m1*m2)/M)
```

```python
# Função auxiliar para calcular o F-Measure adaptada para agrupamento
def F_Measure(MatrizContingencia):
    val = []
    ndotj = MatrizContingencia.sum(axis=0)
    nidot = MatrizContingencia.sum(axis=1)
    for j in range(MatrizContingencia.shape[1]):
        partial = [
            (
                ((MatrizContingencia[i,j]*MatrizContingencia[i,j])/(nidot[i]*ndotj[j]))/
                (MatrizContingencia[i,j]/nidot[i] + MatrizContingencia[i,j]/ndotj[j])
            )
            for i in range(MatrizContingencia.shape[0])
        ]
        
        val.append((ndotj[j]/MatrizContingencia.sum())*max(partial))

    F_Measure_Calculado = sum(val)
    
    return F_Measure_Calculado
```
